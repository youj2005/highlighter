```python
from bs4 import BeautifulSoup
import os
import numpy as np

def remove_items(test_list, item): 
    # using list comprehension to perform the task 
    res = [i for i in test_list if i != item] 
    return res 

def extract_list_texts(html):
    soup = BeautifulSoup(html, 'html.parser')
    
    # Find all <ul> or <ol> elements
    lists = soup.find_all(['li'])
    
    extracted_texts = []
    highlight_masks = []
    
    # Extract text from each list item
    for item in lists:
        if (item.get_text() == ""):
            continue
        try:
            item = str(item).split("<li")[1]
            item = BeautifulSoup("<li" + item, "html.parser")
        except:
            pass
            
        mask = []
        text = []
            
        for text_node in item.find_all(text=True):
            parent = text_node.parent
            is_highlighted = 0
            if parent.name == 'code':
                is_highlighted = 1
            elif parent.name == 'mark':
                if ((parent['class'][0] == 'highlight-teal_background') | (parent['class'][0] == 'highlight-blue_background')):
                    is_highlighted = 2
                elif (parent['class'][0] == 'highlight-yellow_background'):
                    is_highlighted = 3
            elif parent.name == 'li':
                is_highlighted = 0
            else:
                continue
            
            strs = remove_items(text_node.split(" "), "")
            
            mask.extend([is_highlighted] * len(strs))
            text.extend(strs)
        
        if (text):
            extracted_texts.append(text)
            highlight_masks.append(mask)
    
    return extracted_texts, highlight_masks

# Example HTML containing lists with sublists
dir = 'notes'
data = []

for filename in os.listdir(dir):
    filename = os.path.join(dir, filename)
    with open(filename, "r", encoding="utf-8") as f:
        # Parse the HTML
        html = f.read()
        extracted_texts, highlight_masks = extract_list_texts(html)
        
        for i in range(len(extracted_texts)):
            data.append([extracted_texts[i], highlight_masks[i]])
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.to_pickle("data.pkl")

df = pd.read_pickle("data.pkl")
display(df)

print(df[1][0])
```
